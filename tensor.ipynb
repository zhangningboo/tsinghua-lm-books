```python
import torch

t = torch.arange(12)
t
```

```python
t.stride(), t.shape, t.is_contiguous(), t.storage_offset()
```

```python
r = t.reshape(3, 4)  # 原先12个，先分成 3 份， 再每份 4 个
r
```

```python
r.stride(), r.shape, r.is_contiguous(), r.storage_offset()
```

```python
rr = t.reshape(2, 3, 2)
rr
```

```python
rr.stride(), rr.shape, rr.is_contiguous(), rr.storage_offset()
```

```python
c = t.reshape(3, 4).transpose(0, 1)  # 转置
c
```

```python
c.stride(), c.shape, c.is_contiguous(), c.storage_offset()
```

```python
c.view(2, 6) # transpose操作后，内存不连续，无法使用view方法
```

```python
c.contiguous().view(2, 6)  # 先用contiguous方法让内存连续，再用view方法
```
